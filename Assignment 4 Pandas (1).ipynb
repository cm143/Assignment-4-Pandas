```python
import pandas as pd
petal = pd.read_csv("Petal_Data.csv", sep = ",")
sepal = pd.read_csv("Sepal_Data.csv", sep = ",")
```

```python
petal
```

```python
sepal
```

```python
flower = pd.merge(petal,sepal, on = 'sample_id', how = 'inner')
```

```python
flower
```

```python
flower['species_x'].unique()
```

```python
vers_x = flower[flower['species_x'] == 'versicolor']
virg_x = flower[flower['species_x'] == 'virginica']
seto_x = flower[flower['species_x'] == 'setosa']
vers_y = flower[flower['species_y'] == 'versicolor']
virg_y = flower[flower['species_y'] == 'virginica']
seto_y = flower[flower['species_y'] == 'setosa']
```

```python
print('the following are correlations')
#versicolor correlation#
print('versicolor')
print('correlation between vers_x petal_length and vers_x petal_width: ',vers_x['petal_length'].corr(vers_x['petal_width']))
print('correlation between vers_x petal_length and vers_y sepal_length: ',vers_x['petal_length'].corr(vers_y['sepal_length']))
print('correlation between vers_x petal_length and vers_y sepal_width: ',vers_x['petal_length'].corr(vers_y['sepal_width']))
print('correlation between vers_x petal_width and vers_y sepal_width: ',vers_x['petal_width'].corr(vers_y['sepal_length']))
print('correlation between vers_x petal_width and vers_y sepal_width: ',vers_x['petal_width'].corr(vers_y['sepal_width']))
print('correlation between vers_y sepal_length and vers_y sepal_width: ',vers_y['sepal_length'].corr(vers_y['sepal_width']))
#virginica correlation#
print('virginica')
print('correlation between virg_x petal_length and virg_x petal_width: ',virg_x['petal_length'].corr(virg_x['petal_width']))
print('correlation between virg_x petal_length and virg_y sepal_length: ',virg_x['petal_length'].corr(virg_y['sepal_length']))
print('correlation between virg_x petal_length and virg_y sepal_width: ',virg_x['petal_length'].corr(virg_y['sepal_width']))
print('correlation between virg_x petal_width and virg_y sepal_width: ',virg_x['petal_width'].corr(virg_y['sepal_length']))
print('correlation between virg_x petal_width and virg_y sepal_width: ',virg_x['petal_width'].corr(virg_y['sepal_width']))
print('correlation between virg_y sepal_length and virg_y sepal_width: ',virg_y['sepal_length'].corr(virg_y['sepal_width']))
#setosa correlation#
print('setosa')
print('correlation between seto_x petal_length and seto_x petal_width: ',seto_x['petal_length'].corr(seto_x['petal_width']))
print('correlation between seto_x petal_length and seto_y sepal_length: ',seto_x['petal_length'].corr(seto_y['sepal_length']))
print('correlation between seto_x petal_length and seto_y sepal_width: ',seto_x['petal_length'].corr(seto_y['sepal_width']))
print('correlation between seto_x petal_width and seto_y sepal_width: ',seto_x['petal_width'].corr(seto_y['sepal_length']))
print('correlation between seto_x petal_width and seto_y sepal_width: ',seto_x['petal_width'].corr(seto_y['sepal_width']))
print('correlation between seto_y sepal_length and seto_y sepal_width: ',seto_y['sepal_length'].corr(seto_y['sepal_width']))
#mean calculation#
print('the following are average calculations')
print('vers_x petal_length average: ',vers_x['petal_length'].mean())
print('vers_x petal_width average: ',vers_x['petal_width'].mean())
print('vers_y sepal_length average: ',vers_y['sepal_length'].mean())
print('vers_y sepal_width average: ',vers_y['sepal_width'].mean())
print('virg_x petal_length average: ',virg_x['petal_length'].mean())
print('virg_x petal_width average: ',virg_x['petal_width'].mean())
print('virg_y sepal_length average: ',virg_y['sepal_length'].mean())
print('virg_y sepal_width average: ',virg_y['sepal_width'].mean())
print('seto_x petal_length average: ',seto_x['petal_length'].mean())
print('seto_x petal_width average: ',seto_x['petal_width'].mean())
print('seto_y sepal_length average: ',seto_y['sepal_length'].mean())
print('seto_y sepal_width average: ',seto_y['sepal_width'].mean())
#median calculation#
print('the following are median calculations')
print('vers_x petal_length median: ',vers_x['petal_length'].median())
print('vers_x petal_width median: ',vers_x['petal_width'].median())
print('vers_y sepal_length median: ',vers_y['sepal_length'].median())
print('vers_y sepal_width median: ',vers_y['sepal_width'].median())
print('virg_x petal_length median: ',virg_x['petal_length'].median())
print('virg_x petal_width median: ',virg_x['petal_width'].median())
print('virg_y sepal_length median: ',virg_y['sepal_length'].median())
print('virg_y sepal_width median: ',virg_y['sepal_width'].median())
print('seto_x petal_length median: ',seto_x['petal_length'].median())
print('seto_x petal_width median: ',seto_x['petal_width'].median())
print('seto_y sepal_length median: ',seto_y['sepal_length'].median())
print('seto_y sepal_width median: ',seto_y['sepal_width'].median())
#standard deviation calculation#
print('the following are standard deviation calculations')
print('vers_x petal_length standard deviation: ',vers_x['petal_length'].std())
print('vers_x petal_width standard deviation: ',vers_x['petal_width'].std())
print('vers_y sepal_length standard deviation: ',vers_y['sepal_length'].std())
print('vers_y sepal_width standard deviation: ',vers_y['sepal_width'].std())
print('virg_x petal_length standard deviation: ',virg_x['petal_length'].std())
print('virg_x petal_width standard deviation: ',virg_x['petal_width'].std())
print('virg_y sepal_length standard deviation: ',virg_y['sepal_length'].std())
print('virg_y sepal_width standard deviation: ',virg_y['sepal_width'].std())
print('seto_x petal_length standard deviation: ',seto_x['petal_length'].std())
print('seto_x petal_width standard deviation: ',seto_x['petal_width'].std())
print('seto_y sepal_length standard deviation: ',seto_y['sepal_length'].std())
print('seto_y sepal_width standard deviation: ',seto_y['sepal_width'].std())
```

```python
#The species of irises that are most similar are versicolor and setosa since they have the lowest standard deviations in petal width, and share similar low standard deviaitons throughout variables in species x and y#
#The species of irises that are least similar are versicolor and virginica since they share a high standard deviation in petal length and sepal length, and they do not share many positive or negative correlations#
```
